```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
df = load_diabetes()
print(df.DESCR)
```

```python
X = df.data
y = df.target
```

```python
lda = LinearDiscriminantAnalysis(n_components=1)
X_r = lda.fit(X, y).transform(X)
```

Метрики MSE, MAE реализуем сами.

Средняя абсолютная ошибка (MAE) – это среднее абсолютное значение ошибок. Он рассчитывается так: $$\frac{1}{n} \sum_{i=1}^n |y_i - \widehat{y_i}|$$
    
Среднеквадратичная ошибка (MSE) – это среднее значение квадратов ошибок, которое рассчитывается так: $$\frac{1}{n} \sum_{i=1}^n |y_i - \widehat{y_i}|^2$$

```python
def MSE(y: np.array, y_pred: np.array):
    return np.mean((y - y_pred)**2)

    
def MAE(y: np.array, y_pred: np.array):
    return np.mean(np.abs(y - y_pred))
```
